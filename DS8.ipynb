```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
dataset = sns.load_dataset('titanic')

# Show first 5 rows
dataset.head()
```

```python
# Histogram of age
sns.histplot(dataset['age'], bins=10, kde=False)

plt.title("Age Distribution")
plt.show()
```

```python
# Joint plot (scatter)
sns.jointplot(x='age', y='fare', data=dataset, kind='scatter')
```

```python
sns.jointplot(x='age', y='fare', data=dataset, kind='hex')
```

```python
# Bar plot
sns.barplot(x='sex', y='age', data=dataset)

plt.title("Average Age by Gender")
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)

plt.title("Age Variation by Gender")
plt.show()
```

```python
# Count plot for gender
sns.countplot(x='sex', data=dataset)

plt.title("Count of Passengers by Gender")
plt.show()
```

```python
sns.countplot(x='survived', data=dataset)

plt.title("Survival Count")
plt.show()
```

```python
# Box plot (age vs gender)
sns.boxplot(x='sex', y='age', data=dataset)

plt.title("Age Distribution by Gender")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue='survived')

plt.title("Age Distribution by Gender & Survival")
plt.show()
```

```python
# Correlation matrix
corr = dataset.corr(numeric_only=True)

# Heatmap
sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Histogram of fare
sns.histplot(dataset['fare'], bins=10, kde=False)

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```
